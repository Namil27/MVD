````python
patterns = {
    'city_state_zip': r'^[A-Za-z .\-]+,\s*[A-Z]{2}\s*\d{5}$',
    'city_state': r'^[A-Za-z .\-]+,\s*[A-Z]{2}$',
    'state_zip': r'^[A-Z]{2}\s*\d{5}$',
    'state_only': r'^[A-Z]{2}$',
    'zip_only': r'^\d{5}$',
    'city_only': r'^[A-Za-z .\-]+$',
}

def detect_location_format(location):
    if pd.isna(location):
        return 'unknown'
    for name, pattern in patterns.items():
        if re.match(pattern, location.strip()):
            return name
    return 'unknown'

df['location_format'] = df['location'].astype(str).map(detect_location_format)

format_counts = df['location_format'].value_counts()

format_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Форматы значений в столбце location')
plt.xlabel('Формат')
plt.ylabel('Количество')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

#### Задание 1.2

В большей части строк удалось извлечь данные из ячейки столбца `location`, но остались и такие, где это не получилось. Вынесите в отдельный датасет записи, в которых не удалось определить формат записи информации о местоположении.

Над оставшимися строками проведите следующие преобразования:
1. Данные о местоположении разделите на 3 столбца: город, код штата, почтовый индекс.
2. Так как значения в столбце `почтовый индекс` являются пятизначными числами, преобразуйте их в целочисленный формат.

````python
df_unknown_location = df[df['location_format'] == 'unknown'].copy()
df_valid_location = df[df['location_format'] != 'unknown'].copy()

def extract_location_parts(location):
    city = state = zip_code = None
    location = location.strip()

    zip_match = re.search(r'\b\d{5}\b', location)
    if zip_match:
        zip_code = int(zip_match.group())

    state_match = re.search(r'\b[A-Z]{2}\b', location)
    if state_match:
        state = state_match.group()

    if ',' in location:
        city_part = location.split(',')[0]
        city = city_part.strip()
    else:
        parts = location.split()
        for part in parts:
            if not re.fullmatch(r'[A-Z]{2}|\d{5}', part):
                city = part.strip()
                break

    return pd.Series([city, state, zip_code])

df_valid_location[['city', 'state', 'zip']] = df_valid_location['location'].apply(extract_location_parts)

print(df_valid_location[['location', 'city', 'state', 'zip']].head())
````

#### Задание 2.1

В столбце `salary`, как и в `location`, данные не стандатизированы. В нем в основном можно встретить записи в следующих форматах:

* 10.00 - 20.00 &dollar; &sol;hour
* 10,000.00 - 20,000.00 &dollar; &sol;year
* &dollar;10.00 &sol;hour
* &dollar;10,000.00+ &sol;year
* Up to &dollar;20,000.00

Основные проблемы, которые стоит учесть при обработке форматов:
- В каждом из форматов может быть или не быть знака доллара.
- Может отсутствовать информация о периоде, за который указывается зарплата (`/hour` или `/year` в примерах).
- Зарплата может быть указана за месяц (`/month`) и за неделю (`/week`).
- В начале, конце и середине некоторых строк есть лишние пробелы, которые не стоит учитывать.

**Задание:**
1. Определите весь перечень встречающихся форматов и напишите одно или несколько регулярных выражений для их поиска.
2. Покажите на столбчатой диаграмме количество записей, содержащих каждый из форматов. Сделайте выводы.

````python
salary_series = df['salary'].dropna().astype(str).str.strip()

salary_patterns = {
    'range_hour': r'^\$?\s*\d{1,3}(?:[,\d]*)\s*-\s*\$?\s*\d{1,3}(?:[,\d]*)\s*\$?\s*/\s*hour$',
    'range_year': r'^\$?\s*\d{1,3}(?:[,\d]*)\s*-\s*\$?\s*\d{1,3}(?:[,\d]*)\s*\$?\s*/\s*year$',
    'fixed_hour': r'^\$?\s*\d{1,3}(?:[,\d]*)\.?\d*\s*\$?\s*/\s*hour$',
    'fixed_year': r'^\$?\s*\d{1,3}(?:[,\d]*)\.?\d*\+?\s*/\s*year$',
    'upto': r'^Up to\s+\$?\d{1,3}(?:[,\d]*)\.?\d*',
    'monthly': r'^\$?\s*\d{1,3}(?:[,\d]*)\.?\d*\s*/\s*month$',
    'weekly': r'^\$?\s*\d{1,3}(?:[,\d]*)\.?\d*\s*/\s*week$',
}

def classify_salary(s):
    s = s.strip()
    for label, pattern in salary_patterns.items():
        if re.match(pattern, s, re.IGNORECASE):
            return label
    return 'other'

salary_format_series = salary_series.map(classify_salary)
salary_format_counts = salary_format_series.value_counts()

salary_format_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Форматы значений в столбце salary')
plt.xlabel('Формат')
plt.ylabel('Количество записей')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print()
print('Большинство значений в столбце salary имеют нестандартизованный формат,\n'
 'который не удаётся однозначно классифицировать. Наиболее чётко распознаются форматы\n'
  'с указанием годовой зарплаты и выражением “Up to”, что позволяет использовать\n'
   'их для анализа после дополнительной обработки.')
````

````python
def parse_salary(s):
    s = s.strip().replace(',', '')
    min_sal = max_sal = None
    period = None

    if 'hour' in s.lower():
        period = 'hour'
    elif 'year' in s.lower():
        period = 'year'
    elif 'month' in s.lower():
        period = 'month'
    elif 'week' in s.lower():
        period = 'week'

    numbers = re.findall(r'\d+\.\d+|\d+', s)
    numbers = list(map(float, numbers))

    if 'up to' in s.lower() and numbers:
        min_sal = None
        max_sal = numbers[0]
    elif '-' in s and len(numbers) == 2:
        min_sal, max_sal = numbers
    elif numbers:
        min_sal = numbers[0]
        max_sal = None

    return pd.Series([min_sal, max_sal, period])

df_salary = salary_series.to_frame(name='salary')
df_salary[['min_salary', 'max_salary', 'period']] = df_salary['salary'].apply(parse_salary)

group_stats = df_salary.groupby('period')['min_salary'].agg(['mean', 'median'])

group_stats.plot(kind='bar', figsize=(10, 6))
plt.title('Средняя и медианная минимальная зарплата по периодам')
plt.ylabel('Зарплата ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print()
print('Средняя и медианная минимальная зарплата сильно различаются в зависимости от периода,\n'
      'причём годовые значения значительно выше остальных. Это подтверждает необходимость\n'
      'учёта единицы измерения при анализе зарплатных данных.')
````

#### Задание 2.3

Как было отмечено в задании 2.1, период, за который указывается зарплата, может отсутствовать. Восполните этот пробел в данных следующим образом:

1. Определите диапазон изменения (разброс) значений денежной суммы за каждый из встречающихся в данных периодов. Постройте гистограмму для каждого периода.
2. Определите, к среднему значению за какой период ближе всего сумма, у которой период не указан. Укажите для найденный период. В спорных случаях, где сумма одинаково близка к нескольким диапазонам, не выбирайте ничего.

````python
known_periods = df_salary[df_salary['period'].notna()].copy()
unknown_periods = df_salary[df_salary['period'].isna()].copy()

range_by_period = known_periods.groupby('period')['min_salary'].agg(['min', 'max'])
range_by_period['range'] = range_by_period['max'] - range_by_period['min']

for period in known_periods['period'].dropna().unique():
    plt.figure(figsize=(6, 4))
    known_periods[known_periods['period'] == period]['min_salary'].dropna().plot(kind='hist', bins=20)
    plt.title(f'Гистограмма минимальной зарплаты — {period}')
    plt.xlabel('Зарплата ($)')
    plt.ylabel('Количество')
    plt.tight_layout()
    plt.show()

period_means = known_periods.groupby('period')['min_salary'].mean()

def closest_period(val):
    if pd.isna(val):
        return None
    distances = abs(period_means - val)
    if (distances == distances.min()).sum() > 1:
        return None
    return distances.idxmin()

unknown_periods['inferred_period'] = unknown_periods['min_salary'].map(closest_period)

inferred_counts = unknown_periods['inferred_period'].value_counts()

print()
print('Для строк без указанного периода была выполнена попытка восстановления периода на основе сравнения\n'
      'значения зарплаты со средними значениями в каждом известном периоде. В спорных случаях период не присваивался.')
print('\nРаспределение найденных периодов:\n')
print(inferred_counts)
````

Вы заметили, что многие значения фактически одинаковые, но немного по-разному отформатированы? Приведем их к единому формату.

````python
def normalize_job_type(value):
    if pd.isna(value):
        return 'Unknown'

    value = value.lower().strip()

    if 'full' in value:
        if 'intern' in value:
            return 'Full Time Intern'
        return 'Full Time'
    elif 'part' in value:
        if 'intern' in value:
            return 'Part Time Intern'
        return 'Part Time'
    elif 'contract' in value or 'temporary' in value or 'project' in value:
        return 'Contract / Temporary'
    elif 'per diem' in value:
        return 'Per Diem'
    elif 'intern' in value:
        return 'Intern'
    elif 'employee' in value:
        return 'Employee'
    elif 'exempt' in value:
        return 'Exempt'

    return 'Other'

df['job_type_normalized'] = df['job_type'].map(normalize_job_type)
normalized_counts = df['job_type_normalized'].value_counts()

plt.figure(figsize=(10, 6))
normalized_counts.plot(kind='bar')
plt.title('Распределение нормализованных значений в job_type')
plt.xlabel('Тип занятости (нормализовано)')
plt.ylabel('Количество')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print()
print('Значения в столбце job_type были успешно приведены к унифицированным категориям,\n'
      'таким как Full Time, Part Time, Contract / Temporary и т.д., что значительно упростит анализ.')
````

Столбец `job_type` состоит из двух частей: "полнота" занятости (`Full Time`, `Part Time`, `Per Diem`) и "тип" занятости (`Employee`, `Intern`). Кроме того, встречаются и неполные записи.

Задание:
1. Написать одно или несколько регулярных выражений для разделения столбца `job_type` на два: `полнота занятости`, `тип занятости`.
2. Если данные о полноте отсутствуют, внесите значение `Full Time`.
Если отсутствуют данные о типе – внесите `Employee`.
3. Постройте столбчатую диаграмму со стековой группировкой для полноты и типа занятости. Сделайте выводы.

````python
def split_job_type(value):
    if pd.isna(value):
        return pd.Series(['Full Time', 'Employee'])  # по умолчанию

    value = value.lower().strip()

    if 'part' in value:
        fullness = 'Part Time'
    elif 'per diem' in value:
        fullness = 'Per Diem'
    elif 'full' in value:
        fullness = 'Full Time'
    else:
        fullness = 'Full Time'

    if 'intern' in value:
        job_kind = 'Intern'
    elif 'employee' in value:
        job_kind = 'Employee'
    elif 'exempt' in value:
        job_kind = 'Exempt'
    else:
        job_kind = 'Employee'

    return pd.Series([fullness, job_kind])

job_type_split = df['job_type'].apply(split_job_type)
job_type_split.columns = ['job_fullness', 'job_kind']
df[['job_fullness', 'job_kind']] = job_type_split

grouped = df.groupby(['job_fullness', 'job_kind']).size().unstack(fill_value=0)

grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Полнота и тип занятости (job_type)')
plt.xlabel('Полнота занятости')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

print()
print('Значения столбца job_type были успешно разделены на полноту и тип занятости.\n'
      'На графике видно, что доминирует Full Time Employee, но также представлены Intern и Per Diem позиции.')
````
